```python
sheets = pd.read_excel('C:/Users/HP/Downloads/ECOM.xlsx', sheet_name=['customer', 'Sales', 'Product'])
```

```python
top50_customer = sheets['customer'].head(50).to_dict(orient='records') 
 
top50_sales = sheets['Sales'].head(50).to_dict(orient='records') 
 
top50_product = sheets['Product'].head(50).to_dict(orient='records')
```

```python
customer_data = pd.DataFrame(top50_customer) 
print() 
sales_data = pd.DataFrame(top50_sales) 
print() 
product_data = pd.DataFrame(top50_product)
```

```python
customer_data.to_csv('customer_top50.csv', index=0) 
sales_data.to_csv('sales_top50.csv', index=0) 
product_data.to_csv('product_top50.csv', index=0)
```
