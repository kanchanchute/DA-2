# Header

```python
import requests as rq
```

```python
url ="https://www.amazon.in/"

rResp = rq.get(url)
```

```python
dcontent = rResp.content.decode('utf-8')
```

```python
print(dcontent)
```

```python
type(rResp.headers)

rheader = rResp.headers
```

```python
print(rheader)
```

```python
rheader['Date']
```

```python
rheader['Content-Type']
```

```python
userAgent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
```

```python
checkHeader = rq.get(url, headers={'User-Agent' : userAgent})
```

```python
print(checkHeader)
```

```python
checkHeader.request.headers
```

# Params

```python
import requests as rq
```

```python
aUrl = 'https://api.openaq.org/v2/cities'

aParams = {'country': 'AT', 'city': '1020 Wien', 'count': 259518 }

aResp = rq.get(aUrl, params=aParams)
```

```python
print(aResp.json())
```

# Authentication

```python
import requests as rq
```

```python
aUrl = 'https://api.openaq.org/v2/cities'

aAuth = ('city', '1020 Wien')

type(aAuth)
```

```python
content = rq.get(aUrl, auth=aAuth)
content
```
